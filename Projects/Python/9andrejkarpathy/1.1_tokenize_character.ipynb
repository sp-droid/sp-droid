```python
from character_tokenizer import CharacterTokenizer
```

```python
with open("datasets/chosen_dataset.info", "r", encoding="utf-8") as file:
    name = file.read().strip()
dataset = f"datasets/{name}.txt"
with open(dataset, "r", encoding="utf-8") as file:
    text = file.read()
print(text[:1000])

print("Text length in characters:", len(text)/1e6, "M")
```

```python
tokenizer = CharacterTokenizer()
tokenizer.fit(text)
tokenizer.save("intermediate/tokenizer")
```

```python
tokenizer.encode(text)[:100]
```
