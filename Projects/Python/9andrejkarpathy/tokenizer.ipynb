```python
import tokenizers
from tqdm import tqdm
import matplotlib.pyplot as plt
```

```python
# name = "shakespeare"
name = "TinyStoriesV2-GPT4-valid"
```

```python
dataset = f"datasets/{name}.txt"
with open(dataset, "r", encoding="utf-8") as file:
    text = file.read()

print("Length of the text:", len(text))

chars = sorted(list(set(text)))
print("Unique characters:", chars, "Number of unique characters:", len(chars))
```

```python
vocab_sizes = [32, 128, 1024, 2048, 4096, 8192, 16384, 32768]
n_tokens = []
for vocab_size in tqdm(vocab_sizes):
    tokenizer = tokenizers.Tokenizer(tokenizers.models.BPE())
    tokenizer.pre_tokenizer = tokenizers.pre_tokenizers.ByteLevel()
    tokenizer.decoder = tokenizers.decoders.ByteLevel()
    trainer = tokenizers.trainers.BpeTrainer(
        vocab_size=vocab_size,
        special_tokens=["<PAD>", "<UNK>", "<BOS>", "<EOS>"]
    )
    tokenizer.train([dataset], trainer)
    n_tokens.append(len(tokenizer.encode(text).tokens))

plt.plot(vocab_sizes, n_tokens)
plt.xlabel("Vocabulary Size")
plt.ylabel("Number of Tokens")
plt.title(f"{name}: Vocabulary vs Tokens")
plt.grid()
plt.show()
```
